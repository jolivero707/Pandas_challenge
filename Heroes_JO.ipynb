```python
bins = [0,9,14,19,24,29,34,39,100]
group_names =["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
#df = df["Age"]
#df
df["Age Summary"] = pd.cut(df["Age"], bins, labels=group_names)
df
```

```python
df_group = df.groupby("Age Summary")["SN"].count()

avg_purch_price = df.groupby(["Age Summary"])["Price"].mean()

total_purch_value = df.groupby(["Age Summary"])["Price"].sum()

avg_per_person = df.groupby(["Age Summary"])["Price"].sum()/df2.groupby(["Age Summary"])["Age"].count()

Age_summary1 = pd.DataFrame({"Purchase Count" : df_group,
                                     "Average Purchase Price" : avg_purch_price,
                                     "Total Purchase Value" : total_purch_value,
                                     "Avg Total Purchase per Person": avg_per_person})
Age_summary1.round(2)
```

```python
# Top Spenders
Spenders_count = df["SN"].value_counts()
Spenders_mean = df.groupby("SN")["Price"].mean()
Spenders_sum = df.groupby("SN")["Price"].sum()

Spenders_df = pd.DataFrame({"Total count": Spenders_count,
                              "Average Purchase Price": Spenders_mean,
                              "Total Purchase Value": Spenders_sum})
Spenders_df.sort_values(by= ["Total Purchase Value"],ascending=False).head().round(2)
```

```python
# Most Popular Items
df_popular = purchase_data[["Item ID", "Item Name", "Price"]]
df_popular_Items = df_popular.groupby(["Item ID", "Item Name"])["Item Name"].count()
df_popular_price = df_popular.groupby(["Item ID","Item Name"])["Price"].mean()
```

```python
total_popular_value = df_popular.groupby(["Item ID","Item Name"])["Price"].sum()
```

```python
popular_item_summary = pd.DataFrame({"Total Count":df_popular_Items,
                              "Item Price":df_popular_price,
                              "Total Purchase Value": total_popular_value})
```

```python
popular_item_summary.sort_values(by= ["Total Count"],ascending=False).round(2)
```
